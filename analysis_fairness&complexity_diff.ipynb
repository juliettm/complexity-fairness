```python
import pandas as pd
```

```python
complexity_df = pd.read_csv('results/df_complexity_measures.csv', index_col=0)
complexity_df
```

```python
for col in complexity_df.columns:
    complexity_df[col] = abs(complexity_df[col])
complexity_df
```

```python
fairness_df_KN = pd.read_csv('results/combined_quartiles_fairness_KN.csv')
fairness_df_DT = pd.read_csv('results/combined_quartiles_fairness_DT.csv')
fairness_df_LR = pd.read_csv('results/combined_quartiles_fairness_LR.csv')
fairness_df_LR
```

```python
fairness_df_LR = fairness_df_LR[fairness_df_LR['Measure'] != 'std']
fairness_df_LR = fairness_df_LR[['Dataset', 'Att', 'Acc', 'SP', 'EO', 'PP']]
fairness_df_LR['complete_name'] = fairness_df_LR.reset_index().apply(lambda row: f"{row['Dataset']}_{row['Att']}", axis=1).values
fairness_df_LR.set_index('complete_name', inplace=True)
fairness_df_LR.drop(columns=['Dataset', 'Att'], inplace=True)
fairness_df_LR.sort_index(inplace=True)


fairness_df_DT = fairness_df_DT[fairness_df_DT['Measure'] != 'std']
fairness_df_DT = fairness_df_DT[['Dataset', 'Att', 'Acc', 'SP', 'EO', 'PP']]
fairness_df_DT['complete_name'] = fairness_df_DT.reset_index().apply(lambda row: f"{row['Dataset']}_{row['Att']}", axis=1).values
fairness_df_DT.set_index('complete_name', inplace=True)
fairness_df_DT.drop(columns=['Dataset', 'Att'], inplace=True)
fairness_df_DT.sort_index(inplace=True)


fairness_df_KN = fairness_df_KN[fairness_df_KN['Measure'] != 'std']
fairness_df_KN = fairness_df_KN[['Dataset', 'Att', 'Acc', 'SP', 'EO', 'PP']]
fairness_df_KN['complete_name'] = fairness_df_KN.reset_index().apply(lambda row: f"{row['Dataset']}_{row['Att']}", axis=1).values
fairness_df_KN.set_index('complete_name', inplace=True)
fairness_df_KN.drop(columns=['Dataset', 'Att'], inplace=True)
fairness_df_KN.sort_index(inplace=True)
fairness_df_KN
```

```python
fairness_df_KN.rename(columns={'Acc': 'Acc_KN', 'SP': 'SP_KN', 'EO': 'EO_KN', 'PP': 'PP_KN'}, inplace=True)
fairness_df_DT.rename(columns={'Acc': 'Acc_DT', 'SP': 'SP_DT', 'EO': 'EO_DT', 'PP': 'PP_DT'}, inplace=True)
fairness_df_LR.rename(columns={'Acc': 'Acc_LR', 'SP': 'SP_LR', 'EO': 'EO_LR', 'PP': 'PP_LR'}, inplace=True)
fairness_df_KN
```

```python
fairness_df = pd.merge(fairness_df_KN, fairness_df_DT, left_index=True, right_index=True)
fairness_df = pd.merge(fairness_df, fairness_df_LR, left_index=True, right_index=True)
fairness_df
```

```python
for col in fairness_df.columns:
    fairness_df[col] = abs(fairness_df[col])
fairness_df
```

```python
real_data = pd.merge(complexity_df, fairness_df, left_index=True, right_index=True)
real_data.rename(columns={'clustering_coefficient': 'cls_coef'}, inplace=True)
real_data.columns
```

```python
# The rules
rules = [
    {"antecedent": "C2 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "EO_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_LR > 0.1 and EO_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "EO_LR > 0.1"},
    {"antecedent": "C2 > 0.1 and density > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_LR > 0.1 and EO_LR > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_KN > 0.1 and EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_KN > 0.1 and EO_KN > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_DT > 0.1 and PP_DT > 0.1"},
    {"antecedent": "density > 0.1 and C2 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "PP_DT > 0.1"},
    {"antecedent": "density > 0.1 and C2 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "L3 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "L3 > 0.1", "consequent": "EO_LR > 0.1"}
]


# Function to compute rule metrics
def evaluate_rule(data, antecedent, consequent):
    
    antecedent_rows = data.query(antecedent)
    consequent_rows = data.query(consequent)
    both_rows = antecedent_rows.query(consequent)
    antecedent_frequency = len(antecedent_rows)
    consequent_frequency = len(consequent_rows)
    
    if len(antecedent_rows) == 0 or len(consequent_rows) == 0:
        print(antecedent, '->', consequent, ' not found')
        return antecedent_frequency, len(both_rows), antecedent_frequency/len(data),consequent_frequency/len(data), len(both_rows)/len(data),0,0
    else:
        # Calculate metrics
        confidence = (len(both_rows)/len(data)) / (antecedent_frequency/len(data)) if antecedent_frequency > 0 else 0
        lift = confidence / (consequent_frequency/len(data)) if consequent_frequency >0 else 0
        return antecedent_frequency, len(both_rows), antecedent_frequency/len(data),consequent_frequency/len(data), len(both_rows)/len(data), confidence, lift

# Iterating over rules and computing results
results = []
for rule in rules:
    antecedent_frequency, combined_frequency, antecedent_support, consequent_support, support, confidence, lift = evaluate_rule(real_data, rule["antecedent"], rule["consequent"])
    results.append({
        "rule": f"{rule['antecedent']} -> {rule['consequent']}",
        "antecedent_frequency": antecedent_frequency,
        "combined_frequency": combined_frequency,
        "support": support,
        "confidence": confidence,
        "lift": lift
    })

# Convert results to DataFrame for readability
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Define functions to compute the metrics for rule verification
def compute_support(data, conditions):
    """Compute the support for a given condition set."""
    condition_met = data.eval(conditions)
    return condition_met.mean()

def compute_rule_metrics(data, antecedent, consequent):
    """Compute metrics for a rule given antecedent and consequent conditions."""
    antecedent_support = compute_support(data, antecedent)
    consequent_support = compute_support(data, consequent)
    joint_support = compute_support(data, f"({antecedent}) and ({consequent})")
    confidence = joint_support / antecedent_support if antecedent_support > 0 else 0
    lift = confidence / consequent_support if consequent_support > 0 else 0
    return {
        "antecedent_support": antecedent_support,
        "consequent_support": consequent_support,
        "support": joint_support,
        "confidence": confidence,
        "lift": lift
    }

# Define the rules from the prompt
rules = [
    {"antecedent": "C2 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "EO_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_LR > 0.1 and EO_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "EO_LR > 0.1"},
    {"antecedent": "C2 > 0.1 and density > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_LR > 0.1 and EO_LR > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_KN > 0.1 and EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_KN > 0.1 and EO_KN > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "EO_KN > 0.1"},
    {"antecedent": "N1 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "SP_DT > 0.1 and PP_DT > 0.1"},
    {"antecedent": "density > 0.1 and C2 > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "density > 0.1", "consequent": "SP_DT > 0.1"},
    {"antecedent": "C2 > 0.1", "consequent": "PP_DT > 0.1"},
    {"antecedent": "density > 0.1 and C2 > 0.1", "consequent": "SP_KN > 0.1"},
    {"antecedent": "L3 > 0.1", "consequent": "SP_LR > 0.1"},
    {"antecedent": "L3 > 0.1", "consequent": "EO_LR > 0.1"}
]

# Compute the metrics for each rule and store the results
rule_results = []
for rule in rules:
    metrics = compute_rule_metrics(real_data, rule["antecedent"], rule["consequent"])
    rule_results.append({**rule, **metrics})

# Convert results to a DataFrame for better visualization
rule_results_df = pd.DataFrame(rule_results)
rule_results_df
```
